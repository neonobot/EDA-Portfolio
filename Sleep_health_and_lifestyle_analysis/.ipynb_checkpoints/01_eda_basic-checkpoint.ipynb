### Импорты

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Sleep_health_and_lifestyle_dataset.csv', index_col = 'Person ID')
```

```python
df.head()
```

```python
df.dtypes
```

```python
x = df['Occupation']
y = df['Daily Steps']
plt.grid(True, alpha = 0.4, color = 'red')
plt.bar(x, y, color = 'pink')
plt.xticks(rotation = 90)
plt.title('Соотношение кол-ва шагов к роду деятельности')
plt.show()
```

```python
female = df.loc[df['Gender'] == 'Female']
male = df.loc[df['Gender'] == 'Male']

x1 = female['Age']
y1 = female['Sleep Duration']
x2 = male['Age']
y2 = male['Sleep Duration']


#Первый график
plt.subplot(1,2,1) #1 - ряд, 2 - столбца, 1ый график
plt.plot(x1, y1, marker='.', label = 'Female', color = 'violet') #построение графика, marker - задает тип вида точек пересечения x и y

plt.xticks([20, 30, 40, 50, 60, 70], rotation = 0) #настройка тик 

plt.xlabel('Возраст')
plt.ylabel('Продолжительность сна')

plt.grid(True) 

plt.xlim(25,70) #задает диапазон значений графика(можно приблизить график и посмотреть более детально)

plt.legend() # выводит легенду на график



#Второй график
plt.subplot(1,2,2) #1 - ряд, 2 - столбца, 2ой график
plt.plot(x2, y2, marker='*', label = 'Male', color = 'darkblue') #label задает легенду для графика(но не выводит ее на график)

plt.xticks([20, 30, 40, 50, 60, 70], rotation = 0) #настройка тик 

plt.xlabel('Возраст')
plt.ylabel('Продолжительность сна')

plt.grid(True) 

plt.xlim(25,70) #задает диапазон значений графика(можно приблизить график и посмотреть более детально)

plt.legend() # выводит легенду на график


#Общее
plt.suptitle('Графики зависимости возраста и продолжительности сна для женщин и мужчин') #общий заголовок для подграфиков
plt.tight_layout() #автоматически настраивает параметры подграфиков на холсте, чтобы избежать их наложения

plt.show()
```

```python
# Использование figure и axes
age_groups = pd.qcut(df['Age'], 4, labels=['Молодые','Взрослые','Средний возраст','Пожилые']) 
#делит данные на группы с примерно одинаковым количеством элементов

bins = [20, 35, 45, 55, 70]
age_groups2 = pd.cut(df['Age'], bins=bins, labels=['Молодые','Взрослые','Средний возраст','Пожилые']) 
#Создаёт интервалы равной ширины по всему диапазону данных. Количество элементов в каждом интервале может быть разным.

fig = plt.figure(figsize=(7,5))
ax = fig.add_axes([0, 0, 1, 1])
ax.set_title('Гистограмма самых часто встречающихся возрастных групп')
ax.set_xlabel('Возрастная группа')
ax.set_ylabel('Кол-во человек в возрастной группе')
ax.grid(True)
ax.hist(age_groups2)
plt.show()
```

```python
b = df['Daily Steps']
plt.boxplot(b)
plt.grid(True)
plt.title('Диаграмма размаха кол-ва шагов: минимум, первый квартиль (Q1), медиана, третий квартиль (Q3) и максимум')
plt.show()
```

### Обработка пропусков

```python
df.isna().sum()
```

```python
df['Sleep Disorder'].unique()
```

```python
df['Sleep Disorder'] = df['Sleep Disorder'].fillna('No disorders')
```

### Кодирование категориальных признаков

```python
df = pd.get_dummies(df, columns = ['Sleep Disorder'], prefix='', prefix_sep = '', dtype=int,) 
#кодирование 'Sleep Disorder', убрала префиксы. Параметр drop_first не использовала, т.к. удаляет одну из патологий и оставляет их отсутствие.
```

```python
df = df.drop('No disorders', axis=1) #удаляю столбец отстутствия патологий, те у кого по нулям - без патологий сна
```

```python
df.head()
```

```python
df['Insomnia'].dtypes
```

```python
df['Gender'].unique()
```

```python
df = pd.get_dummies(df, columns = ['Gender'], prefix='', prefix_sep = '' , dtype=int, drop_first = True) 
#кодирование 'Gender', убрала префиксы. прописала drop_first 
```

```python
df.head(2)
```

```python
df['BMI Category'].unique() #['Overweight', 'Normal', 'Obese', 'Normal Weight'] - 4 категории
```

```python
df['BMI Category'] = df['BMI Category'].replace('Normal Weight', 'Normal') #  - из 4 категорий делаем 3: Normal, Overweight, Obese
```

```python
df['BMI Category'].unique()
```

```python
df = pd.get_dummies(df, columns = ['BMI Category'], dtype=int, prefix_sep = ' ', drop_first = True) #кодирование 'BMI Category'
```

```python
df.head(2)
```

```python
o = df['Occupation']
sd = df['Sleep Duration']

plt.bar(o, sd)
plt.xticks(rotation = 90)
plt.grid(alpha = 0.45)
plt.title('Кол-во часов сна для профессии')
plt.show()
```

### Разделение категории ‘Blood Pressure’  на две отдельные категории

```python
df['Blood Pressure'].unique()
```

```python
df[['Systolic_pressure','Diastolic_pressure']] = df['Blood Pressure'].str.split('/' ,expand=True).astype('int') 
#expand=True - преобразует в отдельные признаки, если будет false вернет списком
```

```python
df[['Systolic_pressure','Diastolic_pressure']].dtypes
```

```python
df = df.drop('Blood Pressure', axis=1) # удаление первоначального столбца
```

### Кодирование признака 'Occupation'

```python
target = df['Sleep Duration']
```

```python
occupation_target_mean = df.groupby('Occupation')['Sleep Duration'].mean()
```

```python
df['Occupation'] = df['Occupation'].map(round(occupation_target_mean, 1))
```
